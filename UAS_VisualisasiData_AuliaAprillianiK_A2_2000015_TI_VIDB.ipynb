```python
import pandas as pd

# reading the database
data = pd.read_csv("/content/kebutuhan_per_bulan_2.csv")

# printing the top 10 rows
display(data.head(10))
```

```python
pip install matplotlib
```

```python
import pandas as pd
import matplotlib.pyplot as plt

x = data['tahun']
y = data['Propinsi']

plt.scatter(x,y)

plt.plot(x, y, color = 'g', linestyle = 'dashed')

plt.xticks(rotation = 25)
plt.xlabel('Tahun')
plt.ylabel('Propinsi')
plt.title('Kebutuhan per Bulan', fontsize = 12)
plt.grid()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/SupplyKantor.csv")

plt.scatter(data['Jenis'], data['TanggalPesan'], c=data['Harga Satuan'],
            s=data['Jumlah'])

plt.title("Supplay Kantor")
plt.xlabel('Jenis')
plt.ylabel('Tanggal Pesan')

plt.colorbar()

plt.show()
```
